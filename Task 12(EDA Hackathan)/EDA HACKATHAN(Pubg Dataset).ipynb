 <b>Healing:</b>

<b>Boosts:</b>

```python
print(df['boosts'].quantile(0.25))
print(df['boosts'].quantile(0.50))
print(df['boosts'].quantile(0.75))
print(df['boosts'].quantile(0.95))
```
